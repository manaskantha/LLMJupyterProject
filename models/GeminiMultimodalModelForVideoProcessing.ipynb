```python
from http.client import responses

import time
from google import genai
from google.genai import Client
from google.genai.types import Content, CreateCachedContentConfig, HttpOptions, Part, GenerationConfig, GenerateContentConfig

import IPython

client = genai.Client(http_options=HttpOptions())

file_uri = "/Users/manas/Documents/My Files/My Project/vids/IMG_4992.MOV"
myfile = client.files.upload(file=file_uri)

#Wait for the file to be uploaded before sending it to Gemini model
while myfile.state.name == "PROCESSING":
    print('.', end='')
    time.sleep(10) # Wait for 10 seconds before checking again
    myfile = client.files.get(name=myfile.name)

if myfile.state.name == "FAILED":
    raise ValueError(f"File processing failed: {myfile.state.name}")

IPython.display.Video(file_uri, width=640, height=480)

role = """
You are a great digital marketer working on a new video"""

tasks = """
You will add this video to your website and do this you need to complete some tasks.
Please make sure your answer is structured

Tasks:
-What is the title of the video?
-Write a summary of the video
-Generate metadata for this video in JSON format that includes:\
Title, short description, language and company.
"""

content_1 = [myfile, role, tasks]

generation_config_1 = GenerateContentConfig(
    temperature=0.0
)

responses = client.models.generate_content(
    model='gemini-2.5-flash',
    contents=content_1,
    config=generation_config_1,
)

print(responses.text)
```
