```python
import os
from utils import (gemini, gemini_vision, print_multimodal_prompt, gemini_vision_parameters)
import vertexai

from vertexai.generative_models import (GenerativeModel, Image, GenerationConfig)

REGION = "us-central1"
vertexai.init(location=REGION)

model = GenerativeModel('gemini-2.5-flash')
fruits = Image.load_from_file(os.path.join("../data", "fruits_basket.jpeg"))
price = Image.load_from_file(os.path.join("../data", "Price_list.jpg"))

image = [fruits, price]

print_multimodal_prompt(image)

insturction_1 = """
I want to make fruit salad with 3 bananas, two apple, one orange and one kiwi. This is an \
Image of my fruit basket"""

insturction_2 = "This is my price list at the supermarket:"

question = """
Pleasea answer below questions:
- Describe which fruit and how much I have in my fruit basket on the image? \
- Given the fruits in my bowl on the image and the fruits salad recipe, what am I missing? \
- Given the fruits I still need to buy, what would be prices and total cost for these fruits?
"""

content = [
    insturction_1,
    fruits,
    insturction_2,
    price,
    question
]

respone = gemini_vision(content, model)
print(respone, end="")
```
